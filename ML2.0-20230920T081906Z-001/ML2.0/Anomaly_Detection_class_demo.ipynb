```python
# importing libaries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
df = pd.read_csv('AnomalyDetection.csv') 
```

```python
df
```

```python
df['Mileage'] += 10
```

```python
plt.scatter(df['Mileage'],df['Price'], s=5)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
label = EllipticEnvelope(random_state=0, contamination=0.1).fit_predict(df.iloc[:,:2])
df['is_outlier_ee'] = label

# Plotting
plt.scatter(df['Mileage'], df['Price'], c=df['is_outlier_ee'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
less_dense = df.loc[df['Mileage'] <= 80]
```

```python
label = EllipticEnvelope(random_state=0, contamination=0.1).fit_predict(less_dense.iloc[:,:2])
less_dense['is_outlier_ee'] = label



# Plotting
plt.scatter(less_dense['Mileage'], less_dense['Price'], c=less_dense['is_outlier_ee'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```
