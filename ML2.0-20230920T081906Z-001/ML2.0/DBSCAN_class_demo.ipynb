```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('customer_segmentation2.csv')
df.head()
```

```python
X = df.copy()
X.describe()
```

```python
X = X.drop("ID", axis = 1 )
cols = X.columns
```

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
from sklearn.cluster import DBSCAN
dbsc = DBSCAN(eps=0.15,min_samples=5).fit(X)
```

```python
clusters = pd.DataFrame(X, columns=cols)
clusters['dbsc'] = dbsc.labels_
```

```python
clusters['dbsc'].value_counts()
```

```python
import plotly.express as px

fig = px.scatter_3d(clusters, x='n_clicks', y='amount_spent', z='amount_discount', color='dbsc', width=600, height=600)
fig.update_traces(marker=dict(size=2), selector=dict(mode='markers'))
fig.show()
```
