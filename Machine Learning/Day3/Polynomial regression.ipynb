```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
X = np.random.rand(n,1)
```

```python
Y  = 0.7*(X**5)-2.1*(X**4)- 2.3*(X**3)+ 0.2*(X**2)+ 0.3*X+ 0.4*np.random.rand(n,1)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression,Poly
```

```python
model = LinearRegression()
model.fit(X,Y)
output = model.predict(X)

plt.scatter(X,Y)
plt.scatter(X,output)
plt.xlabel("X")
plt.xlabel("Y")
plt.show()
print(model.score(X,Y))
```

```python
X_deg2 = np.hstack([X,X**2])
model = LinearRegression()
model.fit(X_deg2,Y)
output = model.predict(X_deg2)

plt.scatter(X,Y)
plt.scatter(X,output)
plt.xlabel("X")
plt.xlabel("Y")
plt.show()
print(model.score(X_deg2,Y))
```

```python
X_deg3 = np.hstack([X,X**2,X**3])
model = LinearRegression()
model.fit(X_deg3,Y)
output = model.predict(X_deg3)

plt.scatter(X,Y)
plt.scatter(X,output)
plt.xlabel("X")
plt.xlabel("Y")
plt.show()
print(model.score(X_deg2,Y))
```
