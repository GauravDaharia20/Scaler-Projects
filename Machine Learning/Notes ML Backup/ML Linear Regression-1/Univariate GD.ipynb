```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars24-car-price-clean.csv')
```

```python
df
```

```python
X = df['model'].values
Y = df['selling_price'].values
```

```python
def predict(x,weights):
    y_hat = weights[1]*x + weights[0]
    return y_hat

def error(x,y,weights):
    n = x.shape[0]
    total_err = 0
    for i in range(n):
        y_pred = predict(x[i],weights)
        err = (y[i]-y_pred)**2
        total_err += err
    return total_err/n

def gradient(x,y,weights):
    n = x.shape[0]
    
    grads = np.zeros((2,))
    
    for i in range(n):
        
        y_hat = predict(x[i],weights)
        Y = y[i]
        grads[0]+=(-2*(Y-y_hat))
        grads[1]+=(-2*(Y-y_hat)*x[i])
        
    return grads/n
        
def gradient_decent(x,y,epochs=20,lr=0.1):
    weights = np.zeros((2,))
    total_error=[]
    weight_list=[]
    for i in range(epochs):
        grad = gradient(x,y,weights)
        
        errors = error(x,y,weights)
        total_error.append(errors)
        weights[0]-= lr*grad[0]
        weights[1]-= lr*grad[1]
        
        weight_list.append((weights[0],weights[1]))
    
    return weights,weight_list,total_error        
```

```python
weights,weight_list,total_error = gradient_decent(X,Y)
```

```python
plt.plot(list(range(20)),total_error)
plt.xlabel('epochs')
plt.ylabel('errors')
```

```python
plt.plot(np.array(weight_list)[:,0],label='weight0')
plt.legend()
# plt.xlabel('epochs')
# plt.ylabel('weight0')
```

```python
plt.plot(np.array(weight_list)[:,1],label='weight1')
```

```python
fig = plt.figure()
Y_hat = predict(X,weights)
plt.scatter(X,Y)
plt.plot(X,Y_hat,color='orange',label="Prediction")
plt.legend()
plt.show()
```
