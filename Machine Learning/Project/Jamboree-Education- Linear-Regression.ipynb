# VIF Check

```python
from sklearn.model_selection import train_test_split as tts
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

# Standardizing data

```python
def check_vif(df):
    tmp = df.columns
    df = StandardScaler().fit_transform(df)
    df = pd.DataFrame(df,columns=tmp)
    vif_df = pd.DataFrame()
    vif_df['Features'] = df.columns
    vif_df["VIF"] = [round(variance_inflation_factor(df.values, i),2) for i in range(len(df.columns))]
    return vif_df,df
```

```python
import statsmodels.api as sm
X = data.loc[:,'GRE Score':'Research']
Y = data['Chance of Admit']
X = sm.add_constant(X)
X
```

# Spliting data into training and testing

```python
xtrain,xtest,ytrain,ytest = tts(X,Y,test_size=0.25,random_state=0)
```

# Finding Optimum weights using Gradient Descent

```python
def hypothesis(x,weights):
    y_hat = np.dot(x,weights)
    return y_hat

def error(X,Y,weights):
    e=0.0
    m = X.shape[0]
    y_hat = hypothesis(X,weights)
    e = np.sum((Y-y_hat)**2)
    return e/m

def gradient(X,Y,weights):
    y_hat = hypothesis(X,weights)
    grad = np.dot(X.T,(y_hat-Y))
    m = X.shape[0]
    
    return grad/m

def gradient_descent(X,Y,lr=0.1,epoches=1000):
    m = X.shape[1]
    weights = np.zeros((m,))
    error_list = []
    
    for i in range(epoches):
        e = error(X,Y,weights)
        error_list.append(e)
        
        grad = gradient(X,Y,weights)
        
        weights = weights - lr * grad
    return weights,error_list 



def adj_r2_score(r2_score,xtest,ytest):
    num = (1-r2_score)*(len(ytest)-1)
    denom = len(ytest)-xtest.shape[1]-1
    adjr2_sq = 1-(num/denom)
    return adjr2_sq
    
```

```python
weights,error_list = gradient_descent(xtrain,ytrain,lr=0.01)
```

```python
plt.title('error list for training data')
plt.plot(error_list)
plt.xlabel('Epoches')
plt.ylabel('Error')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_gd = hypothesis(xtest,weights)
```

```python
print(f' r2_score : {r2_score(ytest,y_pred_gd)}')
print(f' adj_r2_score : {adj_r2_score(r2_score(ytest,y_pred_gd),xtest,ytest)}')
```

```python
fig = plt.figure()
plt.title('GRE SCORE test data  VS Prediction data')
plt.scatter(xtest['GRE Score'],ytest,label='data')
plt.scatter(xtest['GRE Score'],y_pred_gd,color='orange',label='prediction')
plt.legend()
plt.show()
```

# Assumptions test of linear regression
# Assumption 1: Residuals are independent of each other.
# Assumption 2: Residuals should have a constant variance.

```python
residual_gd = ytest-y_pred_gd
sns.residplot(x=y_pred_gd,y=residual_gd,color='darkorange')
plt.xlabel("y_predict")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

From the above residual plot, we could infer that the residuals didnâ€™t form any pattern. So, the residuals are independent of each other.

And also, the residuals have constant variance. Variance doesnâ€™t seem to increase/decrease constantly with the y_predict value.

# Assumption 3: Residuals are normally distributed
To check whether the residuals are normally distributed from Q-Q plot,distplot

Q-Q plot â†’ quantile-quantile plot
If the residuals are normally distributed, then the Q-Q plot of residuals will be a straight line

```python
from scipy import stats
probplot=sm.ProbPlot(residual_gd,stats.norm,fit=True)
fig=probplot.qqplot(line='45')
plt.title('qqplot')
plt.show()
```

```python
plt.title('Normality check of residuals')
sns.distplot(residual_gd,color='darkorange')
plt.show()
```

# Mean of Residuals

```python
mean_of_residual_gd = np.mean(residual_gd)
print(f'mean_of_residual :{mean_of_residual_gd}')
```

# Test for Hetroskedasticity

```python
sns.residplot(x=ytest,y=residual_gd,color='darkorange')
plt.grid()
plt.ylabel('residual')
plt.show()
```

Above figure shows that there is no sign of hetroskedasity because most of the varience are in range of -0.5 to 0.5
