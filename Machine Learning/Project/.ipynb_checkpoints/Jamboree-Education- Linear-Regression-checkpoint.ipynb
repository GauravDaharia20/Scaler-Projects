```python
def get_histplot(data,title,bins=20):
    plt.title(title)
    plt.grid()
    sns.histplot(data,bins=bins)
    mean = np.mean(data)
    std = np.std(data)
    print(f'mean : {mean} , std : {std}')
    plt.show()
def get_kde_plot(data,title):
    plt.title(title)
    plt.grid()
    sns.kdeplot(data)
    mean = np.mean(data)
    std = np.std(data)
    print(f'mean : {mean} , std : {std}')
    plt.show()
```

```python
count_research = len(data.loc[data['Research']==1])
count_research_gre_gt_320 = len(data.loc[data['Research']==1 & (data['GRE Score']>=320)])
count_research_gre_lt_320 = len(data.loc[data['Research']==1 & (data['GRE Score']<=320)])

count_non_research_gre_gt_320 = len(data.loc[data['Research']==0 & (data['GRE Score']>=320)])
count_non_research_gre_lt_320  = len(data.loc[data['Research']==0 & (data['GRE Score']<=320)])


print(f'researcher count having gre_score > 320 -> {count_research_gre_gt_320} , researcher count having gre_score < 320 -> {count_research_gre_lt_320}')
print(f'non researcher count having gre_score > 320 -> {count_non_research_gre_gt_320} , non researcher count having gre_score < 320 -> {count_non_research_gre_lt_320}')

count_no_research = len(data.loc[data['Research']==0])
print(f'researcher count : {count_research} , no_researcher_count : {count_no_research}')
```

# Data Insights

1. Students with no research background and having gre_Score > 320 have high probability of getting admission.

2. Students with research background and having gre_Score > 320 have high probability of getting admission.

3. Students with no research background have probability of admission in range of 0.4 to 0.8.

4. Students with research background have probability of admission in range of 0.7 to 1.0 .
