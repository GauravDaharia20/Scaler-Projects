```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('knn_imputation.csv')
```

```python
data
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
```

```python
scaler = MinMaxScaler()
df_imp = pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
```

```python
df_imp
```

```python
imputer = KNNImputer(n_neighbors=5)
df_imp = pd.DataFrame(imputer.fit_transform(df_imp),columns=df_imp.columns)
df_imp
```
