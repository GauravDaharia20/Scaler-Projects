```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('multiclass.csv')
```

```python
df
```

```python
sns.histplot(data=df,x='class')
```

```python
X = df.loc[:,'Region':'Delicassen']
Y = np.array(df['class']).reshape(-1,1)
```

```python
X.shape,Y.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X = pca.fit_transform(X)
X.shape
```

```python
df1 = pd.DataFrame(X,columns=['f1','f2'])
df1['y'] = Y
sns.scatterplot(data=df1,x='f1',y='f2',hue='y')
```

```python
from sklearn.model_selection import train_test_split as tts
x_train_val,x_test,y_train_val,y_test = tts(X,Y,test_size=0.25,random_state=0)
x_train,x_val,y_train,y_val = tts(x_train_val,y_train_val,test_size=0.25,random_state=0)
```

```python
print('train',x_train.shape,y_train.shape)
print('test',x_test.shape,y_test.shape)
print('validation',x_val.shape,y_val.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
```

```python
def KNN(X,Y,query_pnt,K=5):
    N = X.shape[0]
    d = X.shape[1]
    # store the euclidean distance and labels 
    dist_labels = [(np.sqrt(np.sum((query_pnt-X[i])**2)),Y[i][0]) for i in range(N)]
    
    # sort distance in incresing order
    dist_labels = np.array(sorted(dist_labels,key = lambda x:x[0]))
    
    #get top k
    dist_label_topk = dist_labels[:K,:]
    
    labels,counts = np.unique(dist_label_topk[:,1],return_counts=True)

    index_max_count = np.argmax(counts)
    
    pred_label = labels[index_max_count]
    
    return pred_label
    
```

```python
KNN(x_train,y_train,x_test,5)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay

y_pred_test = np.zeros((x_test.shape[0],1))
for i in range(x_test.shape[0]):
    y_pred_test[i][0] = KNN(x_train,y_train,x_test[i],K=5)
#print(y_pred_test)
print(classification_report(y_test,y_pred_test))
```

```python
cm = confusion_matrix(y_test,y_pred_test)

disp = ConfusionMatrixDisplay(cm)
disp.plot()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=6,metric='euclidean')
knn.fit(x_train,y_train.ravel())
y_pred = knn.predict(x_test)

print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)

disp = ConfusionMatrixDisplay(cm)
disp.plot()
```
