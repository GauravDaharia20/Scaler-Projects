```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (20, 6)
```

```python
mobile_sales = pd.read_excel('mobilesales.xlsx')
mobile_sales.head()
```

```python
mobile_sales.describe(datetime_is_numeric=True)
```

```python
mobile_sales.shape
```

```python
mobile_sales.set_index('DATE',inplace=True)
```

```python
mobile_sales.head()
```

```python
mobile_sales['Sales'].plot()
```

```python
mobile_sales['Sales'].fillna(method='ffill').plot(label='imputed')
mobile_sales['Sales'].plot(label='original')
plt.legend()
```

```python
mobile_sales['Sales'].interpolate(method='linear').plot(label='imputed')
mobile_sales['Sales'].plot(label='original')
plt.legend()
```

```python
mobile_sales['Sales'] = mobile_sales['Sales'].interpolate(method='linear')
```

```python
mobile_sales['Sales'].isna().sum()
```

```python
mobile_sales['Sales'].hist(bins=50)
```

```python
print(np.percentile(mobile_sales['Sales'],0.5),
np.percentile(mobile_sales['Sales'],99.5))
```

```python
mobile_sales['Sales'].clip(upper=np.percentile(mobile_sales['Sales'],99.5),lower=np.percentile(mobile_sales['Sales'],0.5)).plot(label='clipped')
mobile_sales['Sales'].plot(label='original')
plt.grid()
plt.legend()
plt.show()
```

```python
mobile_sales['Sales'].plot(label='original')
mobile_sales['Sales'].rolling(window=12,center=False).mean().plot(label='smoothen')
plt.legend()
plt.show()
```

```python
mobile_sales['Sales'].plot(label='original')
mobile_sales['Sales'].rolling(window=12,center=True).mean().plot(label='smoothen')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
model = sm.tsa.seasonal_decompose(mobile_sales['Sales'],model='additive')
model.plot()
```
