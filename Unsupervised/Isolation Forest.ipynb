```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv('AnomalyDetection.csv')
```

```python
data
```

```python
clf = IsolationForest(n_estimators=1000,contamination=0.1).fit_predict(data)
```

```python
plt.scatter(x=data['Mileage'],y=data['Price'],c=clf)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf  = LocalOutlierFactor(n_neighbors=25,contamination=0.1).fit_predict(data)
```

```python
plt.scatter(x=data['Mileage'],y=data['Price'],c=clf)
```
