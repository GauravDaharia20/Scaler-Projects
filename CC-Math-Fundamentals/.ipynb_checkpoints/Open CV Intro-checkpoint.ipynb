```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'img.jpeg'
img = cv2.imread(file)
```

```python
h,w,channel = img.shape
print(f'h:{h},w:{w},channel :{channel}')
```

```python
# MAtplotlib uses BGR so we need to convert it to RGB
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
red_channel = np.copy(img)
red_channel[:,:,1]=0
red_channel[:,:,2]=0
plt.imshow(red_channel)
plt.axis('off')
plt.show()
```

```python
green_channel = np.copy(img)
green_channel[:,:,0]=0
green_channel[:,:,2]=0
plt.imshow(green_channel)
plt.axis('off')
plt.show()
```

```python
blue_channel = np.copy(img)
blue_channel[:,:,0]=0
blue_channel[:,:,1]=0
plt.imshow(blue_channel)
plt.axis('off')
plt.show()
```

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(grey_img)
plt.axis('off')
plt.show()
```

```python
grey_img.shape
```

```python
# HUE
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img[:,:,0],cmap='hsv')
plt.axis('off')
plt.show()
```

```python
# Saturation
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img[:,:,1],cmap='hsv')
plt.axis('off')
plt.show()
```

```python
# value
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img[:,:,2],cmap='hsv')
plt.axis('off')
plt.show()
```
