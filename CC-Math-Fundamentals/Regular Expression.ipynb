```python
email_list = []
for word in words:
    if("@" in word):
        email_list.append(word)
```

```python
def mask(s):
    if('@' in s):
        name,domain = s.split('@')
        return (f'{name[0]}###{name[-1]}@{domain}')
    else:
        return (f'{s} is not valid')
```

```python
for email in email_list:
    print(mask(email))
```
