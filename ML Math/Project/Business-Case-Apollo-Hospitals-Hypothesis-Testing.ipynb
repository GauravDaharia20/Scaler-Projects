# Outlier treatment of hospitalization charges feature.

```python
plt.title('Outlier detection of hospitalization charges')
sns.boxplot(data=data,x='hospitalization charges',orient='h')
plt.show()
```

```python
sample = data['hospitalization charges']
outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
sample_outliers = detect_outliers_iqr(sample)
print("Outliers from IQR method: ", sample_outliers)
```

```python
tenth_percentile = np.percentile(sample, 10)
ninetieth_percentile = np.percentile(sample, 90)
# print(tenth_percentile, ninetieth_percentile)
b = np.where(sample<tenth_percentile, tenth_percentile, sample)
b = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
# print("Sample:", sample)
print("New array:",b)
```

```python
plt.title('Outliers after treatment for hospitalization charges')
sns.boxplot(b,orient='h')
plt.show()
```

```python
charges = b
```

# Check for type of skewness for hospitalization charges

Skewness is a measure of the asymmetry of a distribution. This value can be positive or negative.

A negative skew indicates that the tail is on the left side of the distribution, which extends towards more negative values.

A positive skew indicates that the tail is on the right side of the distribution, which extends towards more positive values.

A value of zero indicates that there is no skewness in the distribution at all, meaning the distribution is perfectly symmetrical.


# Check for type of kurtosis for hospitalization charges

The kurtosis of a normal distribution is 3.

If a given distribution has a kurtosis less than 3, it is said to be playkurtic, which means it tends to produce fewer and less extreme outliers than the normal distribution.

If a given distribution has a kurtosis greater than 3, it is said to be leptokurtic, which means it tends to produce more outliers than the normal distribution.

# Performing levene test to check whether severity level has an impact on hospitalization  charges
Assumption for levene test:
The Levene test tests the null hypothesis that all input samples are from populations with equal variances.

H0 : all input samples are from severity populations with equal variances.
    
Ha : all input samples are from severity populations have no equal variances.

alpha = 0.05

```python
alpha = 0.05
test_stats,p_value = levene(severity_zero,severity_one,severity_two,severity_three,severity_four,severity_five,center='mean')
if(p_value <alpha):
    print("Reject the null hypothesis",p_value)
else:
    print("Fail to reject NULL hypothesis",p_value)
```

# Conclusion

1. Severity level is defining well hospitalization charges.

2. We can conclude that severity level has positive impact on hospitalization charges, so with respect to severity level hospitalization charges will also vary.
